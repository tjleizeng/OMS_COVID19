```python
df_real = pd.read_csv("ridership.csv")

nyc_ridership = pd.read_csv("result/nyc_ridership.csv").values[:,1:]
chicago_ridership = pd.read_csv("result/chicago_ridership.csv").values[:,1:]

# Forecasting result
nyc_pred = pd.read_csv("result/nyc_pred.csv").values[:,1:].clip(min=0)
nyc_pred_upper = pd.read_csv("result/nyc_pred_upper.csv").values[:,1:].clip(min=0)
nyc_pred_lower = pd.read_csv("result/nyc_pred_lower.csv").values[:,1:].clip(min=0)

chicago_pred = pd.read_csv("result/chicago_pred.csv").values[:,1:].clip(min=0)
chicago_pred_upper = pd.read_csv("result/chicago_pred_upper.csv").values[:,1:].clip(min=0)
chicago_pred_lower = pd.read_csv("result/chicago_pred_lower.csv").values[:,1:].clip(min=0)

nyc_pred2 = pd.read_csv("result/nyc_pred2.csv").values[:,1:].clip(min=0)
nyc_pred_upper2 = pd.read_csv("result/nyc_pred_upper2.csv").values[:,1:].clip(min=0)
nyc_pred_lower2 = pd.read_csv("result/nyc_pred_lower2.csv").values[:,1:].clip(min=0)

chicago_pred2 = pd.read_csv("result/chicago_pred2.csv").values[:,1:].clip(min=0)
chicago_pred_upper2 = pd.read_csv("result/chicago_pred_upper2.csv").values[:,1:].clip(min=0)
chicago_pred_lower2 = pd.read_csv("result/chicago_pred_lower2.csv").values[:,1:].clip(min=0)

# nyc_pred3 = pd.read_csv("result/nyc_pred3.csv").values[:,1:].clip(min=0)
# nyc_pred_upper3 = pd.read_csv("result/nyc_pred_upper3.csv").values[:,1:].clip(min=0)
# nyc_pred_lower3 = pd.read_csv("result/nyc_pred_lower3.csv").values[:,1:].clip(min=0)

# chicago_pred3 = pd.read_csv("result/chicago_pred3.csv").values[:,1:].clip(min=0)
# chicago_pred_upper3 = pd.read_csv("result/chicago_pred_upper3.csv").values[:,1:].clip(min=0)
# chicago_pred_lower3 = pd.read_csv("result/chicago_pred_lower3.csv").values[:,1:].clip(min=0)
```

```python
nyc_ridership = (nyc_ridership)[365:,:]
chicago_ridership = (chicago_ridership)[365:,:]
```

```python
nyc_fhv_mean, nyc_taxi_mean, nyc_bike_mean = 9842.222, 3440.607, 747.0655
chicago_fhv_mean, chicago_taxi_mean, chicago_bike_mean = 3712.273, 534.7464, 133.4529

nyc_mean = [nyc_fhv_mean, nyc_taxi_mean, nyc_bike_mean]
chicago_mean = [chicago_fhv_mean, chicago_taxi_mean, chicago_bike_mean]
```

```python
x = [datetime(2020, 1, 1) + timedelta(i) for i in range(578)]
x2 = [datetime(2021, 7, 31) + timedelta(i) for i in range(365)]
x3 = [datetime(2021, 8, 15), datetime(2021, 9, 15), datetime(2021, 10, 15), datetime(2021, 11, 15)]
```

```python
fig, axs = plt.subplots(1,3,figsize = (12, 3))
fig.subplots_adjust(hspace=0, wspace=.4)

names = ['NYC_FHV', 'NYC_TAXI', 'NYC_BSS']

for h in range(3):
    
    axs[h].plot(x, np.sum(nyc_ridership[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'C0',lw=.5, label='Real')
    
    axs[h].plot(x2, np.sum(nyc_pred[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'C2',lw=.5, label='No policy')
    axs[h].plot(x2, np.sum(nyc_pred2[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'C1',lw=.5, label='Same policy\n in late 2020')
#     axs[h].plot(x2, np.sum(nyc_pred3[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'C3',lw=1)
    axs[h].fill_between(x2, np.sum(nyc_pred_lower[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, np.sum(nyc_pred_upper[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'green',alpha=0.1)
    axs[h].fill_between(x2, np.sum(nyc_pred_lower2[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, np.sum(nyc_pred_upper2[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'orange',alpha=0.1)
#     axs[h].fill_between(x2, np.sum(nyc_pred_lower3[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, np.sum(nyc_pred_upper3[:,(142*h):(142*h+142)], axis = 1)/2/nyc_mean[h]/71, color = 'red',alpha=0.1)
    if(h==0):
        axs[h].plot(x3[:-1], df_real[names[h]].astype('int').values[:-1]/nyc_mean[h]/71,'-x', color = 'C0',lw=2)
    else:
        axs[h].plot(x3, df_real[names[h]].astype('int').values/nyc_mean[h]/71,'-x', color = 'C0',lw=2)
    
    axs[h].set_ylabel('Trips (%)')
    axs[h].xaxis.set_major_locator(matplotlib.dates.MonthLocator(interval=6))
    axs[h].xaxis.set_major_formatter(matplotlib.dates.DateFormatter("%b'%y"))
    axs[h].tick_params(axis='x', rotation=30)
    axs[h].set_xlim([date(2020, 1, 1), date(2022, 7, 31)])

# axs[0].set_yticks([0,1,2,3,4])
# axs[0].set_yticklabels([0,100,200,300,400])
axs[0].set_yticks([0,0.5,1,1.5,2,2.5])
axs[0].set_yticklabels([0,50,100,150,200,250])

axs[1].set_yticks([0,0.5,1,1.5,2])
axs[1].set_yticklabels([0,50,100,150,200])

axs[2].set_yticks([0,1,2,3,4])
axs[2].set_yticklabels([0,100,200,300,400])

axs[0].grid()
axs[1].grid()
axs[2].grid()
    
axs[0].set_title('NYC FHV')
axs[1].set_title('NYC Taxi')
axs[2].set_title('NYC BSS')

axs[2].legend(loc='center left', bbox_to_anchor=(1, 0.4),ncol = 1)


plt.tight_layout()

plt.savefig('fig/pred_nyc.png', dpi = 300, format = 'png')
plt.savefig('fig/pred_nyc.pdf', dpi = 300, format = 'pdf')
```
