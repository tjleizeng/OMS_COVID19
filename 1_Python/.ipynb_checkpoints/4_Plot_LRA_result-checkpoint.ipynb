```python
from scipy.fft import fft, fftfreq

# Number of sample points
def is_seasonal(y,T=7):

    N = len(y)
    # sample spacing
    T = 1.0 / T
    x = np.linspace(0.0, N*T, N, endpoint=False)
    yf = fft(y)
    yf = 2.0/N * np.abs(yf[0:N//2])
    xf = fftfreq(N, T)[:N//2]
#     plt.plot(xf, yf)
    if np.max(yf[np.where(np.abs(xf-1)<0.1)])>(np.mean(yf)+2*np.std(yf)) or\
                                              np.max(yf[np.where(np.abs(xf-2)<0.1)])>(np.mean(yf)+2*np.std(yf)):
        return True
    else:
        return False
```
